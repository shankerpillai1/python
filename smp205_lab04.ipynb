```python
filename=open('spam1.txt')
total=0.00
numover=0
numoflines=0
for line in filename:
    total=(float(line[line.index('0'):len(line)]))+total
    numoflines=numoflines+1
    if (float(line[line.index('0'):len(line)]))>0.95:
        numover=numover+1
print('The average is', (total/numoflines))
print('There are', numover,'emails that have a spam value of over 95%')
    
```

It would mean that based on the contents of the email and the sender, how similar the email is to emails the user has already flagged as spam previously. The average of 75% means that the computer thinks 3/4 of all the users emails are spam.

```python
filename=open('spam2.txt')
total=0.00
numval=0

for line in filename:
    if line.startswith('From:'):
        sender=line.lstrip('From: ')
    elif line.startswith('To:'):
        reciever=line.lstrip('To: ')
    elif line.startswith('Date:'):
        datetime=line.lstrip('Date: ')
    elif line.startswith('X-DSPAM-Conf'):
        val=(float(line[line.index('0'):len(line)]))
        total=total+val
        numval=numval+1
        if val>0.95:
            print('\n\nSender:',sender,'\nRecepient:',reciever,'\nDate and Time:',datetime,'\nSpam Confidence:',val)

print('\nAverage Value:', (total/numval))
    
    
```
